# Objective - To identify total number of parking spaces available in parking lot.

Today, car parking lot issues have increased due to increase of cars in parking lots.
It has become very challenging to provide a good service to people in parking lots by providing them space for parking in timely manner.
It has also become very challenging to handle all parking availability issues with physical efforts due to the size of parking lots. 
With the help of AI this can me managed in an automated way thereby cutting human efforts and reducing cost related to it.

```python
#setting of working directory
import os
os.getcwd()
```

```python
#reading package and importing required library
from imageai.Detection import ObjectDetection

```

```python
# Using Resnet50 as a pre-trained model 
# and using it with Object Detection package to identify number of cars present in image
# image name "car6" is the image which is given to model
# image name "car66" is the image which successfully shows identified cars in image

execution_path = os.getcwd() # setting path for model to read image

detector = ObjectDetection()
detector.setModelTypeAsRetinaNet()
detector.setModelPath( os.path.join(execution_path , "resnet50_coco_best_v2.0.1.h5")) #using a trained model open cv
detector.loadModel() #loading model
detections = detector.detectObjectsFromImage(input_image=os.path.join(execution_path , "car6.jpg"), output_image_path=os.path.join(execution_path , "car66.jpg"))

#print the object name and its probability of correctly identifying in percentage
for eachObject in detections:
    print(eachObject["name"] , " : " , eachObject["percentage_probability"] )
```

```python
# calculating total numbers of counts of cars
length_key = len(detections)
```

# Assumption - 
Taking total number of car parking space available in image as 42.

Total number of cars found in image was 7. In which one of the car was identifed out of the aspected ratio.
So Total Cars Found in image are 6.

Therefore,
     Total parking space available = (Taking total number of actual car parking space available ) - (Total number of cars found in image)
    
     Total parking space available= 42-6 = 36
        
# Hence, total number of car parking space left in parking lot is for 36 cars.

Possible Improvements

The next possible improvements that can be done over our solution is to bound the image to specific aspect ratio for better detection of objects so that the model dont capture objects out of the specified ratio and exact numbers of cars get detected within the region so that the exact number of free parking spaces available can be determined.
